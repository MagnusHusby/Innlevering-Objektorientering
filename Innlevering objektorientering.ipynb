```python
from pylab import *

h = 10e-5

class Diff:
    def __init__(self, f):
        self.f = f
    
    def framdiff(self, x):
        return (f(x + h) - f(x)) / h
    
    def bakdiff(self, x):
        return (f(x) - f(x - h)) / h
    
    def sentdiff(self, x):
        return (f(x + h) - f(x - h)) / (2 * h)
    
    def firepunktdiff(self, x):
        return (- f(x + 2 * h) + 8 * f(x + h) - 8 * f(x - h) + f(x - 2 * h)) / (12 * h)
```

b)

```python
# En eksempelfunksjon
def f(x):
    return sin(2 * pi * x)

# Definerer en instans med eksempelfunksjonen som funksjon
fder = Diff(f)

# Deriverer eksempelfunksjonen gjennom metodene i klassen for Ã¥ sammenligne
print(fder.framdiff(4))
print(fder.bakdiff(4))
print(fder.sentdiff(4))
print(fder.firepunktdiff(4))
```

```python
# Den analystiske funksjonen for den deriverte av eksempelfunksjonen
def ana_fder(x):
    return 2 * pi * cos(2 * pi * x)

# Definerer definisjonsmengde
x_start = -1
x_slutt = 1
N = 10e5
x = linspace(x_start, x_slutt, N)


# Tomme lister for utdata
ver_analytisk = []
ver_framdiff = []
ver_bakdiff = []
ver_sentdiff = []
ver_firepunktdiff = []

# LÃ¸kke som definerer all utdataen ved Ã¥ appende den deriverte av i i respektive lister
for i in x:
    ver_analytisk.append(ana_fder(i))
    ver_framdiff.append(fder.framdiff(i))
    ver_bakdiff.append(fder.bakdiff(i))
    ver_sentdiff.append(fder.sentdiff(i))
    ver_firepunktdiff.append(fder.firepunktdiff(i))
    
# Plotter de forskjellige verdiene opp mot hverandre
plot(x, ver_analytisk, color = "red")
title("Analystisk representasjon av den deriverte")
xlabel("x")
ylabel("f(x)")
grid()
show()

plot(x, ver_framdiff, color = "blue")
title("Numerisk representasjon av den deriverte ved bruk av framoverdifferanse")
xlabel("x")
ylabel("f(x)")
grid()
show()

plot(x, ver_bakdiff, color = "cyan")
title("Numerisk representasjon av den deriverte ved bruk av bakoverdifferanse")
xlabel("x")
ylabel("f(x)")
grid()
show()

plot(x, ver_sentdiff, color = "navy")
title("Numerisk representasjon av den deriverte ved bruk av sentraldifferanse")
xlabel("x")
ylabel("f(x)")
grid()
show()

plot(x, ver_firepunktdiff)
title("Numerisk representasjon av den deriverte ved bruk av firepunktsdifferanse")
xlabel("x")
ylabel("f(x)")
grid()
show()
```
